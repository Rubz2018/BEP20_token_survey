```python
fig1 = px.line(nms_price_vlm_df, x='date', y=['open','high','low','close'],title="Historical Price for NMS token", height=400, width=800)
fig1.show()
```

```python
cols = ['date','open', 'high', 'low', 'close']
nms_price_df = nms_price_vlm_df[cols]
nms_price_df = nms_price_df.set_index('date') 
diff_nms_price_df = nms_price_df.diff()
diff_nms_price_df
```

```python
fig2 = px.line(diff_nms_price_df, x= diff_nms_price_df.index, y=['open','high','low','close'],title="Historical Price change for NMS token", height=400, width=800)
fig2.show()
```

```python
fig3 = px.line(nms_price_vlm_df, x= 'date', y='volume_usd',title="Daily transacted volume for NMS token", height=400, width=800)
fig3.show()
```

```python
vol = ['date','volume_usd']
nms_vol_df = nms_price_vlm_df[vol]
nms_vol_df = nms_vol_df.set_index('date') 
diff_nms_vol_df = nms_vol_df.diff()
diff_nms_vol_df
```
